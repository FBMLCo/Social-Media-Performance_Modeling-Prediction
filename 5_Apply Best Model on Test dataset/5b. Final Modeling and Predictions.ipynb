## Create a train, validation and test datasets (from the main Train set of data)

### Create X and y arrays

```python
# Create X_train array
X_train = svm_train_60k.drop(["All_impact_bins"], axis=1)
X_train = np.array(X_train)
X_train.shape
```

```python
# Create y_train array
y_train = svm_train_60k["All_impact_bins"]
y_train = np.array(y_train)
y_train.shape
```

```python
pd.value_counts(y_train, normalize=True)
```

```python
# Create X_test array
X_test = svm_master_test.drop(["All_impact_bins"], axis=1)
X_test = np.array(X_test)
X_test.shape
```

```python
# Create y_test array
y_test = svm_master_test["All_impact_bins"]
y_test = np.array(y_test)
y_test.shape
```

### Apply Best Model selected in previous steps: SVM Linear Kernel (C = 0.1, decision function shape = 'ovo')

```python
# Create SVM pipeline with the same parameters as the ones used for the the selected Best Model candidate
pipe_svm = Pipeline(
    [
        ("scaler", StandardScaler()),  # with standardization StandardScaler()
        ("PCA", PCA(n_components=200)),  # 200 components to explain 95% of the variance
        (
            "SVM",
            SVC(kernel="linear", C=0.1, decision_function_shape="ovo", random_state=0),
        ),
    ]
)
```

```python
# Fit SVM
model_svm = pipe_svm.fit(X_train, y_train)
```

```python
# Make prediction on X_valid dataset
y_pred_svm = pipe_svm.predict(X_test)
```

```python
# Confusions report
target_names = ["class 0", "class 1", "class 2"]
print(classification_report(y_test, y_pred_svm, target_names=target_names))
```

### Conclusions
> * It appears that the model selected and trained (SVM with Linera Kernel) is performing well with a good ability to generalize
* We can notice a high f1 score across classes which was not the case at the beginning of the modeling phase.
